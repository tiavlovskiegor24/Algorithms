```python
def plot_coo_matrix(m,color = "black",ms = 0.1):
    if not isinstance(m, sparse.coo_matrix):
        m = sparse.coo_matrix(m)
    fig = plt.figure(figsize=(8,8))
    ax = fig.add_subplot(111, axisbg='white')
    ax.plot(m.col, m.row, 's', color=color, ms=ms)
    ax.set_xlim(0, m.shape[1])
    ax.set_ylim(0, m.shape[0])
    ax.set_aspect('equal')
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.invert_yaxis()
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    return ax
```

# Global mean passage time

```python
from scipy.linalg import eig

def GMFPT_theory (A,weighted=True) :
    """
    According to the theory of Lin et al., 2012, the global mean first passage
    time can be calculated by finding the eigenspectrum of the Laplacian matrix
    of the graph. This function calculates the GMFPT from their formula, for the
    graph described by the adjacency matrix A, to all sites. Optional parameter
    'weighted' allows for the choice of having the same quantity but weighted
    with the stationary distribution.
    """
    N = A.shape[0]
    d = np.sum(A,axis=1)
    E = np.sum(d)/2.
    L = np.diag(d) - A
    L_eigs = eigh(L)
    sortidx = np.argsort(L_eigs[0])
    l = np.array([L_eigs[0][i].real for i in sortidx])
    #print l
    v = np.array([L_eigs[1][:,i].real for i in sortidx])
    T = np.zeros(N)
    dv = np.dot (v,d)
    #print dv
    if not weighted :
        for j in xrange(N) :
            for i in range(1,N) :
                T[j] += 1.0/l[i] * (2*E*v[i,j]**2 - v[i,j]*dv[i])
            #print T[j]
        return float(N)/(N-1.0) * T
    else :
        for j in xrange(N) :
            for i in range(1,N) :
                dvi = v[i,j]*dv[i]
                T[j] += 1.0/l[i]*((2*E)**2*v[i,j]**2 - 2*v[i,j]*2*E*dvi - dvi**2)
        return T/(2*E)
```

```python
def GMFPT_theory_1(A,weighted=True) :
    """
    According to the theory of Lin et al., 2012, the global mean first passage
    time can be calculated by finding the eigenspectrum of the Laplacian matrix
    of the graph. This function calculates the GMFPT from their formula, for the
    graph described by the adjacency matrix A, to all sites. Optional parameter
    'weighted' allows for the choice of having the same quantity but weighted
    with the stationary distribution.
    """
    N = A.shape[0]
    d = np.sum(A,axis=1)
    E = np.sum(d)/2.
    L = np.diag(d) - A
    l,v = np.linalg.eigh(L)
    sortidx = np.argsort(l)
    l = l[sortidx]
    #print l
    v = v[:,sortidx]
    T = np.zeros(N)
    #print v.shape,d.shape
    dv = np.dot (v.T,d)
    #print dv
    if not weighted :
        #for j in xrange(N) :
            #T[j] = np.dot(l[1:]**-1,2.0*E*v[j,1:]**2 - v[j,1:]*dv[1:])
            #print T[j]
        T = np.dot(2.0*E*v[:,1:]**2 - v[:,1:]*dv[1:],l[1:]**-1)
        
        return float(N)/(N-1.0) * T
    else :
        for j in xrange(N) :
            for k in xrange(1,N) :
                dvi = v[j,k]*dv[k]
                T[j] += 1.0/l[k]*((2*E)**2*v[j,k]**2 - 2*v[j,k]*2*E*dvi - dvi**2)
        return T/(2*E)
```
